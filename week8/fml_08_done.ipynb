## 🔎 When is **Accuracy** unsuitable?

![meme02](https://github.com/rasvob/VSB-FEI-Fundamentals-of-Machine-Learning-Exercises/blob/master/images/fml_08_meme_02.jpg?raw=true)

# Take a look at this [article](https://builtin.com/data-science/precision-and-recall) and [wiki](https://en.wikipedia.org/wiki/Precision_and_recall) about precision and recall
* 💡 Beware the fact that as we increase precision, we decrease recall and vice-versa.

* **Precision** is the ratio of correctly predicted positive observations to the total predicted positive observations.
    * The question that precision answer is of all passengers that labeled as survived, how many actually survived?
    * High precision relates to the low false positive rate.


* **Recall** is the ratio of correctly predicted positive observations to the all observations in actual class - yes.
    * The question recall answers is: Of all the passengers that truly survived, how many did we label?

* **F1 Score** is the harmonic mean of Precision and Recall. Therefore, this score takes both false positives and false negatives into account.
    * Intuitively it is not as easy to understand as accuracy, but F1 is usually more useful than accuracy, especially if you have an uneven class distribution.
    * 💡 **F1 works best if false positives and false negatives have similar cost**
        * If the cost of false positives and false negatives are very different, it’s better to look at both Precision and Recall.

$ConfMatrix = \begin{bmatrix}
TP & FN\\
FP & TN
\end{bmatrix}$

## Let's take a look at an example:
$|1| = 10$

$|0| = 90$

$M = \begin{bmatrix}
1 & 9\\
0 & 90
\end{bmatrix}$

$Accuracy = \frac{TP+TN}{TP+TN+FP+FN} = \frac{1+90}{1+90+0+9} = \frac{91}{100} = 0.91$

$Precision = \frac{TP}{TP+FP} = \frac{1}{1+0} = 1$

$Recall = \frac{TP}{TP+FN} = \frac{1}{1+9} = \frac{1}{10} = 0.1$

$F1-Score = 2 \cdot \frac{Precision \cdot Recall}{Precision+Recall} = 2 \cdot \frac{1 \cdot 0.1}{1+0.1} = 2 \cdot \frac{0.1}{1.1} = 0.09$

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap="Blues", fmt="g")
plt.show()

tn, fp, fn, tp = cm.ravel()

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * (precision * recall) / (precision + recall)

accuracy, precision, recall, f1
```

# Can we improve our evaluation process?
* Lets try **cross-validation** process for the decision tree model
    * https://scikit-learn.org/stable/modules/cross_validation.html
* **What is the difference between Pure and Stratified K-Fold?**
    * 💡 Take a look at the survival ratio

```python
kfold = KFold(n_splits=5)
score = []
for train_index, test_index in kfold.split(X_train, y_train):
  X_train_fold, X_test_fold = X_train.iloc[train_index, :], X_train.iloc[test_index, :]
  y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]

  clf = DecisionTreeClassifier(random_state=13)
  clf.fit(X_train_fold, y_train_fold)
  y_pred = clf.predict(X_test_fold)
  score.append(f1_score(y_test_fold, y_pred))

np.mean(score), np.min(score), np.max(score)
```

# 📊 ML algorithms usually have hyper-parameters which change the behaviour of the model
* It is usually a good idea to check documentation 😊
    * https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier
* For the *DecisionTreeClassifier* a `max_depth` or `min_samples_split` are pretty important

* 🎯 The goal of the hyper-parameter tuning is to investigate the effect of the parameters on the model and ideally make the model better with setting the right parameters
    * Some models are more sensitive to parameters settings

## 🚀 Let's tune the `max_depth` first

```python
scores_global = []

for max_depth in range(2, 25):
    skf = StratifiedKFold(n_splits=5)
    scores = list()
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        clf = DecisionTreeClassifier(random_state=13)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        scores.append(f1_score(y_test, y_pred))
    scores_global.append((max_depth, np.mean(scores)))

scores_global[:10]
```

```python
df_res = pd.DataFrame.from_records(scores_global, columns=['max_depth', 'f1'])
df_res.tail()
```

## Now we can plot the data
* 🔎 Which setting is the best?

```python
sns.lineplot(data = df_res, x='max_depth', y='f1')
```

```python
df_res.sort_values(by='f1', ascending=False).head(5)
```

## 🚀 Now we can continue with `min_samples_split`

```python
scores_global = []
for x in np.arange(2, 50):
    skf = KFold(n_splits=5)
    scores = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        clf = DecisionTreeClassifier(min_samples_split = x, random_state = 13)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        scores.append(f1_score(y_test, y_pred))
    scores_global.append((x, np.mean(scores)))

scores_global[:10]
```

```python
df_res = pd.DataFrame.from_records(scores_global, columns=['min_samples_split', 'f1'])
df_res.tail(10)
```

```python
sns.lineplot(data = df_res, x='min_samples_split', y='f1')
```

### Which setting is the best?

```python
df_res.sort_values(by='f1', ascending=False).head(5)
```

## 🎯 So now we can try the combination of the best parameter-wise settings

```python
skf = KFold(n_splits=5)
scores = []
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    clf = DecisionTreeClassifier(max_depth = 8, min_samples_split = 12, random_state = 13)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    scores.append(f1_score(y_test, y_pred))

np.mean(scores), np.min(scores), np.max(scores)
```

## Alternatively we can tune both parameters at once 😇

```python
scores_global = []
for d in range(2, 25):
    for split in range(2, 50):
        skf = KFold(n_splits=5)
        scores = []
        for train_index, test_index in skf.split(X, y):
            X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]
            clf = DecisionTreeClassifier(max_depth = d, min_samples_split = split, random_state = 13)
            clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)
            scores.append(f1_score(y_test, y_pred))
        scores_global.append((d, split, np.mean(scores)))
```

# SVC (Support Vector Classifier)

Support Vector Classifier (SVC) je klasifikační model založený na metodě **Support Vector Machines (SVM)**. SVC se snaží najít optimální **hyperrovinu**, která odděluje data do dvou různých tříd s maximálním marginem (vzdáleností) mezi třídami.




```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svc_default = SVC(random_state=13)
svc_default.fit(X_train_scaled, y_train)
y_pred_svc_default = svc_default.predict(X_test_scaled)
f1_svc_default = f1_score(y_test, y_pred_svc_default)

dt_clf = DecisionTreeClassifier(random_state=13)
dt_clf.fit(X_train, y_train)
y_pred_dt = dt_clf.predict(X_test)
f1_dt_default = f1_score(y_test, y_pred_dt)

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.01, 0.1, 1, 10],
    'kernel': ['rbf']
}

grid_search = GridSearchCV(SVC(random_state=13), param_grid, scoring='f1', cv=5)
grid_search.fit(X_train_scaled, y_train)
```

### Shrnutí výsledků

- **SVC (Default)**: Výchozí SVC model dosáhl F1 skóre **0.82**, což je nejlepší výsledek z testovaných modelů.
- **Decision Tree (Default)**: Výchozí Decision Tree model má nižší F1 skóre **0.76**, což ukazuje na horší výkon ve srovnání s SVC.
- **SVC (Tuned Best Params)**: Laděný SVC model s parametry (`C=1`, `gamma=0.1`, `kernel='rbf'`) dosáhl F1 skóre **0.73**, což je nižší než výchozí SVC. Ladění v tomto případě nepřineslo zlepšení a možná vedlo k přetrénování.

#### Závěr
Výchozí SVC model se ukázal jako nejlepší, což naznačuje, že ladění parametrů ne vždy zaručuje lepší výsledky.


```python
best_params_svc = grid_search.best_params_
best_f1_svc = grid_search.best_score_

f1_scores = {
    "SVC (Default)": f1_svc_default,
    "Decision Tree (Default)": f1_dt_default,
    "SVC (Tuned Best Params)": best_f1_svc
}

f1_scores_df = pd.DataFrame(list(f1_scores.items()), columns=["Model", "F1 Score"])
best_params_column = []
for model in f1_scores_df["Model"]:
    best_params_column.append(best_params_svc) if model == "SVC (Tuned Best Params)" else best_params_column.append("-")

f1_scores_df["Best SVC Parameters"] = best_params_column
f1_scores_df
```

  - Rozptyl F1 skóre je menší, což znamená konzistentnější výkon napříč jednotlivými částmi dat.
  - Medián F1 skóre je lehce nižší než u Tuned SVC, což však kompenzuje stabilita modelu.

#### Závěr
Tuned SVC má potenciál vyššího výkonu než Decision Tree, ale za cenu vyšší variability. Výběr modelu závisí na preferenci mezi stabilitou (Decision Tree) a možným dosažením vyššího výkonu (Tuned SVC).


```python
scores_svc = grid_search.cv_results_['mean_test_score']

kf = KFold(n_splits=5, shuffle=True, random_state=13)
scores_dt = []

for train_index, test_index in kf.split(X_train, y_train):
    X_train_fold, X_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]
    dt_clf_fold = DecisionTreeClassifier(random_state=13)
    dt_clf_fold.fit(X_train_fold, y_train_fold)
    y_pred_fold = dt_clf_fold.predict(X_test_fold)
    scores_dt.append(f1_score(y_test_fold, y_pred_fold))

plt.figure(figsize=(12, 6))
sns.boxplot(data=[scores_svc, scores_dt])
plt.xticks([0, 1], ["Tuned SVC", "DecisionTreeClassifier"])
plt.ylabel("F1 Score")
plt.title("Comparison of F1 Scores from Cross-Validation")
plt.show()
```

#### Závěr
Tuned SVC model má lepší schopnost klasifikovat vzorky správně ve srovnání s Decision Tree, což se projevuje vyšší hodnotou AUC. Pokud je důležité minimalizovat počet chybně klasifikovaných příkladů, SVC je lepší volbou.
